1. **Descriptive Statistics**
   - **Correlation Matrix**: Compute pairwise correlations between numerical variables to understand linear relationships.
   - **Covariance Matrix**: Similar to the correlation matrix but also accounts for the scale of variables.
   - **Summary Statistics**: Use descriptive statistics (mean, median, standard deviation, etc.) for each variable to understand their distribution.

2. **Visualization Techniques**
   - **Scatter Plot Matrix (Pair Plot)**: Visualizes all pairwise relationships between numerical variables. Use seaborn's `pairplot` to see scatter plots along with histograms.
   - **Heatmap**: Visualizes the correlation matrix or results from a crosstab. A heatmap can quickly show patterns or areas of high correlation.
   - **Box Plot**: Displays the distribution of numerical data and can be grouped by categorical variables to compare distributions.
   - **Violin Plot**: Combines a box plot and a density plot to show the distribution of a dataset. Useful for comparing distributions across groups.
   - **3D Scatter Plot**: Useful for visualizing the relationship between three numerical variables.
   - **Parallel Coordinates Plot**: Visualizes high-dimensional data by plotting each feature on a separate axis and drawing lines to connect data points.
   - **Cluster Map**: A heatmap with hierarchical clustering, useful for identifying clusters of similar variables.

3. **Combinatorial Approaches**
   - **Crosstab with Heatmap**: Combine crosstab (contingency table) with a heatmap to visualize the relationship between two categorical variables.
   - **Scatter Plot with Regression Line**: Add a regression line to scatter plots to understand the strength and direction of relationships between variables.
   - **Facet Grid**: A matrix of plots, typically used for visualizing multiple relationships between variables by faceting on different categories.
   - **Joint Plot**: Combines scatter plots with histograms and kernel density estimates to show bivariate distributions.
   - **Pair Plot with Histograms**: A scatter plot matrix that includes histograms or KDE plots along the diagonal to show the univariate distribution of each variable.
