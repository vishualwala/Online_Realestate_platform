```python
def fillNa_price(row):
    if pd.isna(row['PRICE']):
        row['PRICE'] = row['Price_per_sqft']* row['AREA']
    return row['PRICE']
```

```python
df['PRICE'] = df.apply(fillNa_price, axis=1)
```

```python
df['PRICE'].isnull().sum()
```

```python
def fillNa_price_per_sq_ft(row):
    if pd.isna(row['Price_per_sqft']):
        row['Price_per_sqft'] = row['PRICE']*10000/ row['AREA']
    return row['Price_per_sqft']
```

```python
df['Price_per_sqft'] = df.apply(fillNa_price_per_sq_ft,axis = 1)
```

```python
def feature_luxury(row):
    try:
        features_str = row['FEATURES']

        features_list = [int(x) for x in features_str.split(',')]

        return sum(features_list)
        
    except ValueError as e:
        return np.nan
        
    except Exception as e:       
        return np.nan

df['feature_luxury'] = df.apply(feature_luxury, axis=1)
```

```python
def amenity_luxury(row):
    try:
        amenity_str = row['AMENITIES']

        amenity_list = [int(x) for x in amenity_str.split(',')]

        return sum(amenity_list)
        
    except ValueError as e:
        return np.nan
        
    except Exception as e:       
        return np.nan

df['amenity_luxury'] = df.apply(amenity_luxury, axis=1)
```

```python


# Convert string representations of dictionaries to actual dictionaries
df['MAP_DETAILS'] = df['MAP_DETAILS'].apply(lambda x: ast.literal_eval(x) if isinstance(x, str) else x)

# Safely extract LATITUDE and LONGITUDE as floats
df['LATITUDE'] = df['MAP_DETAILS'].apply(lambda x: float(x.get('LATITUDE')) if isinstance(x, dict) and 'LATITUDE' in x else None)
df['LONGITUDE'] = df['MAP_DETAILS'].apply(lambda x: float(x.get('LONGITUDE')) if isinstance(x, dict) and 'LONGITUDE' in x else None)
```

```python
def landMark(row):
    try:
        landmark = []
        for i in ast.literal_eval(row):
            landmark.append(i['text'])
        return landmark
        
    except Exception as e:       
        return np.nan


df['FORMATTED_LANDMARK_DETAILS'] = df['FORMATTED_LANDMARK_DETAILS'].apply(landMark)
```

```python
import numpy as np

def get_locality_name(row):
    try:
        row = ast.literal_eval(row)
        return row['LOCALITY_NAME']
    except (KeyError, TypeError):  # KeyError for missing key, TypeError for non-dict
        return np.nan

df['location'] = df['location'].apply(get_locality_name)
```
