```python
import pandas as pd
import random
```

```python
df = pd.read_csv('flatData.csv')
```

```python
pd.set_option('display.max_columns', None)
```

```python
columns_to_drop = ['FORMATTED_LANDMARK_DETAILS']
df.drop(columns=[col for col in columns_to_drop if col in df.columns], axis=1, inplace=True)
```

```python
df['PROPERTY_TYPE'] = df['PROPERTY_TYPE'].apply(lambda x: x.title() if isinstance(x, str) else x)
```

```python

facing_directions = ['North', 'North-East', 'East', 'South-East', 'South', 'South-West', 'West', 'North-West']
yes_no = [True, False]
water_supply_options = ['Borewell', 'Municipal']

restaurants = ['Oh! Calcutta', '6 Ballygunge Place', 'Peter Cat', 'Mainland China', 'Arsalan', 'Flurys', 'Mocambo', 
               'Barbeque Nation', 'Nizam\'s', 'Aminia', 'Shiraz Golden Restaurant', 'Bhojohori Manna', 
               'Kewpie’s', 'Bohemian', 'Peshawri', 'Baan Thai', 'Eden Pavilion', 'Tamarind', 'Santa’s Fantasea', 
               'The Bridge']
railway_stations = ['Howrah Station', 'Sealdah Station', 'Shalimar Station', 'Santragachi Station', 
                    'Ballygunge Junction', 'Dumdum Junction', 'Park Circus Station', 'Tollygunge Station', 
                    'Bidhan Nagar', 'Kolkata Station', 'Patipukur Station', 'Chitpur Terminal', 
                    'Belghoria Station', 'Liluah Station', 'Baghajatin Station', 'Garia Station', 
                    'Barasat Junction', 'Barrackpore Station', 'Jadavpur Station', 'New Barrackpore Station']
malls = ['South City Mall', 'Quest Mall', 'City Centre 1', 'Forum Mall', 'Acropolis Mall', 'Mani Square', 
         'Diamond Plaza', 'Lake Mall', 'Avani Riverside Mall', 'Axis Mall', 'Downtown Mall', 'Shoppers Stop', 
         'City Centre 2', 'Hiland Park Mall', 'Metropolis Mall', 'Fort Knox', 'Gariahat Mall', 
         'Rosedale Plaza Mall', 'Space Town Mall', 'E-Mall']

schools = ['La Martiniere', 'St. Xavier’s School', 'South Point', 'Don Bosco', 'Delhi Public School', 
           'Loreto Convent', 'Modern High School', 'Mahadevi Birla World Academy', 'Birla High School', 
           'Calcutta International School', 'Garden High School', 'The Heritage School', 
           'The Future Foundation School', 'Lakshmipat Singhania Academy', 'Ashok Hall Girls’ Higher Secondary School', 
           'St. James\' School', 'Welland Gouldsmith School', 'Julien Day School', 'Calcutta Girls\' High School', 
           'Adamas International School']
colleges = ['Presidency College', 'Jadavpur University', 'Calcutta University', 'St. Xavier’s College', 
            'Bethune College', 'Lady Brabourne College', 'Shri Shikshayatan College', 'Scottish Church College', 
            'Asutosh College', 'Loreto College', 'Bangabasi College', 'Goenka College of Commerce', 
            'Maulana Azad College', 'Seth Anandram Jaipuria College', 'Bhawanipur Education Society College', 
            'Surendranath College', 'Gokhale Memorial Girls\' College', 'Victoria Institution', 
            'South Calcutta Girls\' College', 'Muralidhar Girls\' College']
hospitals = ['Apollo Gleneagles', 'Fortis', 'Ruby General', 'AMRI', 'Woodlands Hospital', 'CMRI', 'Belle Vue Clinic', 
             'Desun Hospital', 'Peerless Hospital', 'Bhagirathi Neotia Woman & Child Care Centre', 
             'Narayana Multispeciality Hospital', 'Eden Clinic', 'Medica Superspecialty Hospital', 
             'Tata Medical Center', 'Ramakrishna Mission Seva Pratishthan', 'Sparsh Diagnostic Centre', 
             'ILS Hospital', 'GD Hospital', 'Barasat Hospital', 'Bengal Oncology Centre']
markets = ['New Market', 'South City Mall', 'Gariahat Market', 'Mani Square', 'Quest Mall', 'City Centre', 
           'Diamond Plaza', 'Howrah AC Market', 'Bara Bazar', 'Hatibagan Market', 'Lansdowne Market', 
           'Chowringhee Square', 'Spencer\'s Galleria', 'Star Mall', 'Ganguly Bagan Bazaar', 
           'Park Street Market', 'Esplanade Market', 'Salt Lake Market', 'Dharmatala Market', 'Tollygunge Bazaar']

metros = ['Park Street Metro', 'Dumdum Metro', 'Esplanade Metro', 'Rabindra Sadan Metro', 
          'Kalighat Metro', 'Maidan Metro', 'Jatin Das Park Metro', 'Rabindra Sarobar Metro', 
          'Central Metro', 'Chandni Chowk Metro', 'Shyambazar Metro', 'Belgachia Metro', 
          'Netaji Metro', 'Masterda Surya Sen Metro', 'Kavi Nazrul Metro', 'Girish Park Metro', 
          'Mahatma Gandhi Road Metro', 'Kavi Subhash Metro', 'Phoolbagan Metro', 'Salt Lake Sector V Metro']


def generate_distance():
    return round(random.uniform(0.5, 10), 2)


def generate_additional_data(row):
    return {
        "Facing_Direction": random.choice(facing_directions),
        "Loan_Availability": random.choice(yes_no),
        "Estimated_Monthly_EMI": round(random.uniform(30000, 100000), 2) if random.choice(yes_no) else None,
        "Maintenance_Fees": random.randint(2000, 8000),
        "Property_Tax": random.randint(8000, 25000),
        "Stamp_Duty_Registration_Costs": round(random.uniform(3.0, 7.0), 2),
        "Nearest_Schools": {random.choice(schools): generate_distance()},
        "Nearest_Colleges": {random.choice(colleges): generate_distance()},
        "Nearest_Hospitals": {random.choice(hospitals): generate_distance()},
        "Nearest_Markets": {random.choice(markets): generate_distance()},
        "Nearest_Public_Transport": {random.choice(metros): generate_distance()},
        "Nearest_Restaurants": {random.choice(restaurants): generate_distance()},
        "Nearest_Railway_Stations": {random.choice(railway_stations): generate_distance()},
        "Nearest_Malls": {random.choice(malls): generate_distance()},
    
        "Swimming_Pool": random.choice(yes_no),
        "Playground": random.choice(yes_no),
        "RERA_Registration_Number": random.randint(100000, 999999),
        "24x7_Security": random.choice(yes_no),
        "Visitor_Parking": random.choice(yes_no),
        "Intercom_Facility": random.choice(yes_no),
        "Power_Backup": random.choice(yes_no),
        "Water_Supply": random.choice(water_supply_options),
        "Pet_Friendly": random.choice(yes_no),
        "Fire_Safety_Installed": random.choice(yes_no)
    }

additional_data = [generate_additional_data(row) for _, row in df.iterrows()]

df_additional = pd.DataFrame(additional_data)

df_combined = pd.concat([df, df_additional], axis=1)


df_combined.to_json('FlatData.json', orient='records', indent=4)

print("Data has been successfully saved as 'structured_property_data.json'.")
```
